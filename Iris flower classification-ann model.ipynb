### step 1: import libraries

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import load_model

# Save model and scaler
import joblib
```

### step 2 :load dataset

```python
iris=load_iris()
X=iris.data
Y=iris.target
```

```python
# create dataframe
df=pd.DataFrame(X,columns=iris.feature_names)
```

```python
# add  target column
df["Target"]=Y
```

```python
df.head()
```

### step 3: preprocessing

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

```python
print(df.info())
```

### step 4: separate input and output

```python
x=df.drop("Target",axis=1)
y=df["Target"]
```

### step 5: label encoding & one-hot encoding 

```python
# Label encoding
le=LabelEncoder()
y=le.fit_transform(y)
# one-hot encoding 
y_encoded=to_categorical(y)
print(y_encoded[:5])
```

### step6:normalize features

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
print(x_scaled[:5])
```

### step 7:train_split_test

```python
x_train,x_test,y_train,y_test=train_test_split(
    x_scaled,
    y_encoded,
    test_size=0.2,
    random_state=42
)
print(x_train.shape)
print(x_test.shape)
```

### step 8:build the model

```python
model=Sequential()
# hidden layer 1
model.add(Dense(10,input_shape=(4,),activation="relu"))

# hidden layer 2
model.add(Dense(8,activation="relu"))

# output layer
model.add(Dense(3,activation="softmax"))
```

### step 9:compile the model

```python
model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)
```

### step 10 :train the model

```python
history=model.fit(
    x_train,
    y_train,
    epochs=50,
    batch_size=8,
    validation_split=0.2
)
```

### step 11: evaluate the model

```python
loss,accuracy =model.evaluate(x_test,y_test)

print("loss:",loss)
print("accuracy:",accuracy)
```

### step12:prediction

```python
y_pred=model.predict(x_test)

# convert predicted values
y_pred_classes=np.argmax(y_pred,axis=1)

# convert actual values
y_test_classes=np.argmax(y_pred,axis=1)

print("predicted values:")
print(y_pred_classes[:10])

print("Actual values:")
print(y_test_classes[:10])
```

### step 13:accuracy score

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test_classes,y_pred_classes)
print("Accuracy score:",acc)
```

### step 14:classification report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_classes,y_pred_classes))
```

### Step 15: Save Model and Scaler

```python
# Save ANN Model
model.save("iris_ann_model.keras")
# Save Scaler
joblib.dump(scaler, "iris_scaler.pkl")
print("Model and Scaler Saved Successfully")
```

### Step 16: Load Saved Model and Scaler

```python
loaded_model = load_model("iris_ann_model.keras")

loaded_scaler = joblib.load("iris_scaler.pkl")
```

### Step 17: Test with Unseen Sample

```python
# sample flower
unseen_sample = np.array([
    [5.1, 3.5, 1.4, 0.2]
])
```

```python
# scale data
unseen_scaled = loaded_scaler.transform(unseen_sample)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_iris

iris = load_iris()

species_names = iris.target_names

label_encoder = LabelEncoder()
label_encoder.fit(species_names)
```

```python
# Predict
prediction = loaded_model.predict(unseen_scaled)

predicted_class = np.argmax(prediction)

predicted_species = label_encoder.inverse_transform(
    [predicted_class]
)

print("Predicted Flower Species:",
      predicted_species[0])
```

### Conclusion

The ANN model was successfully developed to classify Iris flowers into Setosa, Versicolor, and Virginica. After preprocessing, feature scaling, and training with TensorFlow/Keras, the model achieved high classification accuracy. The results demonstrate that Artificial Neural Networks are effective for multiclass classification tasks and can accurately predict flower species from unseen data.
